### 2 - Les départements et leur région

<br>

Le fichier a été téléchargé, puis déposé sur GitHub, pour pouvoir y accéder à l'aide d'un url. \
On récupère les données du fichier HTML qu'on enregistre dans un data frame. \
La première ligne du fichier HTML est le nom des colonnes, on l'utilise donc. On ne renomme pas la première colonne car on ne l'utilisera pas.

```python
# URL GitHub pour le fichier CSV
url = "https://raw.githubusercontent.com/ahina0/projet_outinfo/main/regions-et-departements-francais-converted.html"

# Télécharger le fichier HTML depuis GitHub
response = requests.get(url)
response.raise_for_status()


soup = BeautifulSoup(response.text, 'html.parser')

# step 2: identify the tags containing the data
# assume the data is in <tr> tags with <td> for each column
rows = soup.find_all('tr')

# step 3: extract and structure the data
data = []
for row in rows:
    cells = row.find_all('td')
    if len(cells) == 3:  # Number, Department, Region
        num = cells[0].get_text(strip=True)        # Number
        dept = cells[1].get_text(strip=True)       # Department
        region = cells[2].get_text(strip=True)     # Region
        data.append([num, dept, region])

# step 4: create a DataFrame with the extracted data
dept_region = pd.DataFrame(data)
dept_region.columns = dept_region.iloc[0]
dept_region = dept_region[1:].reset_index(drop=True)
dept_region
```

On rappelle que le but du projet étant de manipuler des données, on décide de ne pas conserver les colonnes de la commune, du département et de la région. On va maintenant créer nous même une colonne postal_code à partir de la colonne "adresse".

```python
#function to have the postal code 
def extraire_code_postal(adresse):
    import re
    # Repérer 5 chiffres consécutifs dans l'adresse
    match = re.search(r"\d{5}", str(adresse))
    return match.group(0) if match else None

# Ajouter la colonne "postal_code" à la table existante
professionnels_sante["postal_code"] = professionnels_sante["adresse"].apply(extraire_code_postal)

# Afficher le DataFrame mis à jour
professionnels_sante.head()
```

Maintenant, on crée une colonne "dep_code" à partir de notre colonne "costal_code" en sélectionnant les deux premier chiffres de chaque code postal.

```python
#on prend en compte les départements à 3 chiffres
departements_speciaux = {"971", "972", "973", "974", "975", "976", "977", "978", "984", "986", "987", "988", "989"}

#fonction pour extraire le numéro du département
def extraire_dep_code(postal_code):
    if postal_code is None:
        return None
    if postal_code[:3] in departements_speciaux:
        return postal_code[:3]
    return postal_code[:2]


professionnels_sante["dep_code"] = professionnels_sante["postal_code"].apply(extraire_dep_code)
professionnels_sante.head()
```

On veut créer, à partir de notre table "professionnels_sante", une nouvelle table "professionnels_nb_dep" qui contient dans la première colonne les départemets qui apparaissent dans la colonne "dep_code" de "professionnels_sante", et dans la deuxième colonne le nombre de fois où chaque département apparait dans la table "professionnels_sante". Cela nous permettra d'obtenir le combre de médecins par département en 2024 (puis que ces données datent de décembre 2024). 

```python
professionnels_nb_dep = professionnels_sante["dep_code"].value_counts().reset_index()
professionnels_nb_dep.columns = ["dep_code", "count"]

#on trie par ordre croissant
professionnels_nb_dep = professionnels_nb_dep.sort_values(by="dep_code").reset_index(drop=True)

# Afficher la nouvelle table
print(professionnels_nb_dep)
```

```python
# Comparer les données de 2023 et 2024 (fusion des tables)
comparaison = pd.merge(
    professionnels_nb_dep,
    statistiques,
    left_on="dep_code",
    right_on="Code",
    how="inner"
)

# Renommer les colonnes pour simplifier la manipulation
comparaison.rename(columns={
    "Médecin généraliste (en nombre) 2023": "Médecins généralistes 2023"
}, inplace=True)

comparaison.rename(columns={
    "count": "Médecins généralistes 2024"
}, inplace=True)

# Tracer un graphique comparatif
mp.figure(figsize=(15, 8))
bar_width = 0.35
indices = range(len(comparaison))

# Histogramme pour les médecins généralistes 2023
mp.bar(indices, comparaison["Médecins généralistes 2023"], width=bar_width, label="Médecins généralistes 2023")

# Histogramme pour les médecins généralistes 2024 (décalé légèrement pour éviter la superposition)
mp.bar(
    [i + bar_width for i in indices],
    comparaison["Médecins généralistes 2024"],
    width=bar_width,
    label="Médecins généralistes 2024"
)

# Ajouter les légendes, titres et axes
mp.xlabel("Départements")
mp.ylabel("Nombre")
mp.title("Comparaison des médecins généralistes par département (2023 vs 2024)")
mp.xticks([i + bar_width / 2 for i in indices], comparaison["dep_code"], rotation=45, fontsize=7)
mp.legend()

# Afficher le graphique
mp.tight_layout()
mp.show()
```

On compare ces données sur une carte afin de pouvoir faire une analyse plus poussée.

```python


# Charger les données GeoJSON depuis GitHub
geojson_url = "https://raw.githubusercontent.com/gregoiredavid/france-geojson/master/departements.geojson"
geo_df = gpd.read_file(geojson_url)

# Liste des départements d'outre-mer à exclure
outre_mer = {"971", "972", "973", "974", "975", "976", "977", "978", "984", "986", "987", "988", "989"}

# Filtrer pour inclure uniquement la France métropolitaine
geo_df = geo_df[~geo_df["code"].isin(outre_mer)]

# Fonction pour créer la carte interactive
def create_medical_map(selected_year):
    # Vérifier que l'année est valide
    if selected_year not in [2023, 2024]:
        raise ValueError("Année invalide. Choisissez 2023 ou 2024.")
    
    # Choisir la colonne à afficher en fonction de l'année
    colonne = f"Médecins généralistes {selected_year}"
    
    # Fusionner les données avec les géométries
    geo_data = geo_df.merge(comparaison, left_on="code", right_on="dep_code", how="left")
    
    # Créer la carte centrée sur la France
    medical_map = folium.Map(location=[46.603354, 1.888334], zoom_start=6)
    
    # Ajouter les départements avec le dégradé de couleur
    folium.Choropleth(
        geo_data=geo_data,
        data=comparaison,
        columns=["dep_code", colonne],
        key_on="feature.properties.code",
        fill_color="YlGnBu",
        fill_opacity=0.7,
        line_opacity=0.2,
        nan_fill_opacity=0,
        legend_name=f"Médecins généralistes ({selected_year})"
    ).add_to(medical_map)
    
    # Ajouter des info-bulles pour chaque département
    folium.GeoJson(
        geo_data,
        tooltip=folium.GeoJsonTooltip(
            fields=["nom", colonne],
            aliases=["Département:", "Nombre de médecins:"],
            localize=True
        )
    ).add_to(medical_map)
    
    # Afficher la carte
    display(HTML(medical_map._repr_html_()))

# Widget interactif pour choisir l'année
interact(
    create_medical_map,
    selected_year=widgets.SelectionSlider(
        options=[2023, 2024],
        description='Année:',
        orientation='horizontal',
        layout={'width': '600px'}
    )
)
```

Cependant, ce qui compte réellement, c'est le ratio entre le nombre de médecins et la population totale par département.

Pour cela, il nous faut les données de population par 2023. On télécharge donc des données en .json.

```python
# URL GitHub pour le fichier JSON
url = "https://raw.githubusercontent.com/ahina0/projet_outinfo/main/demographyref-france-pop-legale-departement.json"

# Télécharger le fichier JSON depuis GitHub
response = requests.get(url)
response.raise_for_status()

# Charger les données dans un DataFrame
pop_2023 = pd.read_json(response.text)

# Afficher les premières lignes du DataFrame
print(pop_2023.head())
```

Maintenant, on calcule le ratio entre le nombre de médecins généralistes 2023 par département et la population 2023 par départements.

```python
# Vérifier et corriger les colonnes
statistiques["Code"] = statistiques["Code"].apply(lambda x: x[0] if isinstance(x, list) else x)
pop_2023["dep_code"] = pop_2023["dep_code"].apply(lambda x: x[0] if isinstance(x, list) else x)

# Convertir en chaînes pour éviter les erreurs de fusion
statistiques["Code"] = statistiques["Code"].astype(str)
pop_2023["dep_code"] = pop_2023["dep_code"].astype(str)

# Fusionner les deux tables sur le code département
merged_data = statistiques.merge(
    pop_2023,
    left_on="Code",
    right_on="dep_code",
    how="left"
)

# Calculer le ratio médecins/population en pourcentage
merged_data["Ratio médecins/population"] = (
    merged_data["Médecin généraliste (en nombre) 2023"] / merged_data["dep_pop_tot"]
) * 100

# Afficher les premières lignes pour vérifier le résultat
print(merged_data[["Code", "dep_code", "Médecin généraliste (en nombre) 2023", "dep_pop_tot", "Ratio médecins/population"]].head())
```

Affichons les cartes des médecins par département et du ratio par département en 2023 : 

```python
# Charger le GeoJSON directement depuis GitHub
url_geojson = "https://raw.githubusercontent.com/gregoiredavid/france-geojson/master/departements.geojson"
geo_df = gpd.read_file(url_geojson)

# Liste des départements d'outre-mer à exclure
outre_mer = {"971", "972", "973", "974", "975", "976", "977", "978", "984", "986", "987", "988", "989"}

# Filtrer pour inclure uniquement la France métropolitaine
geo_df = geo_df[~geo_df["code"].isin(outre_mer)]

# Fonction pour afficher deux cartes : nombre de médecins et ratios en pourcentage
def afficher_cartes():
    # Fusionner les données statistiques avec les géométries
    carte_data = geo_df.merge(merged_data, left_on="code", right_on="Code", how="left")

    # Créer la figure avec deux sous-graphiques
    fig, axes = mp.subplots(1, 2, figsize=(20, 15))

    # Carte 1 : Nombre de médecins généralistes en 2023
    ax1 = carte_data.plot(
        column="Médecin généraliste (en nombre) 2023",
        cmap="YlGnBu",
        legend=True,
        edgecolor="black",
        linewidth=0.5,
        ax=axes[0]
    )
    ax1.set_title("Nombre de médecins généralistes par département (2023)", fontsize=14)
    ax1.set_axis_off()

    # Ajouter des annotations pour chaque département (carte 1)
    for x, y, label in zip(carte_data.geometry.centroid.x, carte_data.geometry.centroid.y, carte_data["Médecin généraliste (en nombre) 2023"]):
        axes[0].text(x, y, str(label) if not pd.isna(label) else "", fontsize=6, ha="center", color="black")

    # Carte 2 : Ratio médecins/population en pourcentage
    ax2 = carte_data.plot(
        column="Ratio médecins/population",
        cmap="YlGnBu",
        legend=True,
        edgecolor="black",
        linewidth=0.5,
        ax=axes[1]
    )
    ax2.set_title("Ratio Médecins généralistes / Population (2023) [%]", fontsize=14)
    ax2.set_axis_off()

    # Ajouter des annotations pour chaque département (carte 2)
    for x, y, label in zip(carte_data.geometry.centroid.x, carte_data.geometry.centroid.y, carte_data["Ratio médecins/population"]):
        axes[1].text(x, y, f"{label:.2f}%" if not pd.isna(label) else "", fontsize=6, ha="center", color="black")

    # Ajuster la mise en page
    mp.tight_layout()
    mp.show()

# Afficher les deux cartes
afficher_cartes()
```
